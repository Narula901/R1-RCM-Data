```python
def status(row):
    Status = row['Status1']
    
    mapping = {
        "Transacted by Others": "Pending with Ops",
        "Transacted by TB": "Pending with TB",
        "Transacted by BOT": "Pending with BOT"
    }
    
    if Status in mapping:
        return mapping[Status]
    else :
        return Status
    
```

```python

import pandas as pd

df = completion_status.copy()


# Ensure the key input columns exist; if not, create them to avoid KeyErrors.
for col in ['Key_SC_Epic_EI_FC', 'Account ID', 'Key_CBS', 'Key_completion_status',
            'Final Transaction', 'TS', 'CS', 'FT']:
    if col not in df.columns:
        df[col] = pd.NA

# Normalize the "Final Transaction" source column if FT is supposed to mirror "Final Transaction"
# If your 'FT' column is the state you want to use, comment out the line below.
#df['FT'] = df['Final Transaction']

# Pre-create the target columns so .loc assignment works cleanly
for target in ['Tranasction status', 'Completion status', 'Final Transaction']:
    if target not in df.columns:
        df[target] = pd.NA

# ----- Conditions -----
cond_epic_key_notx = df['Key_SC_Epic_EI_FC'].notna() & (df['Final Transaction'] == 'Not Transacted')
cond_account_notx  = df['Account ID'].notna()         & (df['Final Transaction'] == 'Not Transacted')
cond_cbs_tx        = df['Key_CBS'].notna()            & (df['Final Transaction'] == 'Transacted')
cond_key_comp      = df['Key_completion_status'].notna()
penstate = df['Entity Group'] == 'Pennstate'
swedish = df['Entity Group'] == 'Swedish'

# ----- Updates -----
# 1) If Key_SC_Epic_EI_FC present AND FT == 'Not Transacted'
df.loc[cond_epic_key_notx, ['Tranasction status', 'Completion status']] = [
    'EPIC Adj Approval pending', 'EPIC Adj Approval pending'
]

# 2) Else if Account ID present AND FT == 'Not Transacted'
# (Will apply to rows NOT already matched by the first rule automatically due to order)
mask2 = cond_account_notx & ~cond_epic_key_notx
df.loc[mask2, ['Tranasction status', 'Completion status']] = [
    'EPIC Adj Approval pending', 'EPIC Adj Approval pending'
]

# 3) Else if Key_CBS present AND FT == 'Transacted'
df.loc[cond_cbs_tx, ['Tranasction status', 'Completion status']] = [
    'Transacted by BOT', 'Completed'
]

# 4) Else if Key_completion_status present → copy TS/CS/FT into outputs
mask4 = cond_key_comp & ~cond_epic_key_notx & ~mask2 & ~cond_cbs_tx
df.loc[mask4, 'Tranasction status'] = df.loc[mask4, 'TS']
df.loc[mask4, 'Completion status']   = df.loc[mask4, 'CS']
df.loc[mask4, 'Final Transaction']   = df.loc[mask4, 'FT']

#5)
mask5 = swedish
df.loc[mask5, ['Completion status', 'Tranasction status', 'Final Transaction']] = ['Pending with Ops', 'Not worked not transacted', 'Not Transacted']


# Everything else is left unchanged.
```
