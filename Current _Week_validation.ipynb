```python
def validate_row(row):
    group = row['Group']
    PN  = row['Project Name']
    FGN = row['FacilityGroupName']
    

    if group != 24 and pd.isnull(row['EncounterInsuranceBalance U']):
        return "Couldn't fetch the data"  
        
    elif (
    row['EncounterInsuranceBalance U'] == 0 and
    row['EncounterOpenBalance U'] == 0 and
    row['EncounterPatientBalance U'] == 0
    ):
        return "Transacted"
    
    elif group == 0:   # Actually Blank in Group column 
        return "Validation Mapping Error"
    
    elif group == 1 and row['EncounterOpenBalance U'] != row['EncounterOpenBalance']:
        return "Transacted"
    elif group == 2 and row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance']:
        return "Transacted"
    elif group == 3 and row['EncounterInsuranceBalance U'] == 0:
        return "Transacted"
    elif group == 4 and row['EncounterOpenBalance U'] == 0:
        return "Transacted"
    elif group == 5 and row['EncounterPatientBalance U'] == 0:
        return "Transacted"
    elif group == 6 and row['EncounterPatientBalance U'] == 0:
        return "Transacted"
    elif group == 7 and (row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] 
                         or row['CurrentPPC'] != row['Current Payer Plan Code']):
        return "Transacted"
    elif group == 8 and (
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or 
        isinstance(row['ActivityCode U'], str) and "Write" in row['ActivityCode U']
    ):
        return "Transacted"
    elif group == 9 and (
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or 
        row['EncounterInsuranceBalance U'] == 0
    ):
        return "Transacted"
    elif group == 10 and isinstance(row['ActivityCode U'], str) and row['ActivityCode U'].startswith("Write"):
        return "Transacted"
    elif group == 11 and row['CurrentMajoryPayer'] != row['Current Major Payor']:
        return "Transacted"
    elif group == 12 and (
        row['Current Financial Class'] != row['CurrentFinancialClass U'] or 
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance']
    ):
        return "Transacted"
    elif group == 13 and (
        row['CurrentMajoryPayer'] != row['Current Major Payor'] or 
        row['EncounterOpenBalance'] != row['EncounterOpenBalance U']
    ):
        return "Transacted"
    elif group == 14 and (
        row['CurrentMajoryPayer'] != "Medicare"
    ):
        return "Transacted"
    elif group == 15 and (
        row['EncounterInsuranceBalance U'] > 0 
    ):
        return "Transacted"
    elif group == 16 and row['EncounterPatientBalance'] != row['EncounterPatientBalance U']:
        return "Transacted"
    elif group == 17 and (
        row['EncounterInsuranceBalance U'] > 0 or  
        isinstance(row['ActivityCode U'], str) and "Account updated for same day follow up" in row['ActivityCode U']
    ):
        return "Transacted"

    elif group == 19 and row['CurrentPPC'] != row['Current Payer Plan Code']:
        return "Transacted"

    elif group == 20 and (row['CurrentMajoryPayer'] != row['Current Major Payor'] and row['CurrentPPC'] != row['Current Payer Plan Code']):
         return "Transacted"

    elif group == 21 and (
    (isinstance(row['ActivityCode U'], str) and "Account updated for same day follow up" in row['ActivityCode U']) or
    (isinstance(row['ActivityCode2'], str) and "Account updated for same day follow up" in row['ActivityCode2']) or
    (isinstance(row['ActivityCode3'], str) and "Account updated for same day follow up" in row['ActivityCode3']) or
    row['EncounterInsuranceBalance U'] > 0 or
    row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
    row['EncounterOpenBalance U'] == 0
    ):
        
        return "Transacted"

    elif group == 23 and (
        row['EncounterOpenBalance U'] == 0 or 
        row['EncounterInsuranceBalance U'] == 0):
    
        return "Transacted"

    elif group == 24 and (
        row['Primary Payer Plan Code'] != row['Secondary Payer Plan Code']):
        return "Transacted"

    elif group == 25 and (
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] and 
        row['EncounterInsuranceBalance U'] == 0
    ):
        return "Transacted"

    
    elif group == 26 and (
        row['EncounterOpenBalance U'] > 0
        
    ):
        return "Transacted"

    elif group == 27 and (
       row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
       row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance']
    ):
        return "Transacted"

    elif group == 28 and FGN == "Lifepoint" and (
    row['EncounterInsuranceBalance U'] == 0 or
    row['EncounterOpenBalance U'] != row['EncounterOpenBalance']
    ):
        return "Transacted"

    elif group == 28 and FGN != "Lifepoint" and row['EncounterOpenBalance U'] == 0:
        return "Transacted"

    elif group == 29 and (
    row['EncounterPatientBalance U'] == 0 or 
    isinstance(row['ActivityCode U'], str) and 
    "One week review required" in row['ActivityCode U']
    ):
        return "Transacted"

    elif group == 30 and (
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
        row['EncounterPatientBalance'] != row['EncounterPatientBalance U']
    ):
        return "Transacted"

    elif group == 31 and (
        row['EncounterOpenBalance U'] == 0 or
        row['EncounterInsuranceBalance U'] == 0
    ):
        return "Transacted"

    elif group == 32 and (
        row['EncounterTotalCharges U'] > 0
    ):
        return "Transacted"

    elif group == 33 and (
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or
        row['CurrentMajoryPayer'] != row['Current Major Payor']
    ):
        return "Transacted"

    elif group == 34 and (
        row['Current Financial Class'] != row['CurrentFinancialClass U']
    ):
        return "Transacted"

    elif group == 35 and (
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance']
    ):
        return "Transacted"

    elif group == 36 and (
        row['EncounterOpenBalance U'] <= 0 or
        row['EncounterInsuranceBalance U'] == 0 
        
    ):
        return "Transacted"

    elif group == 37 and (
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or 
        isinstance(row['ActivityCode U'], str) and "Unable to process claim until next day" in row['ActivityCode U']
    ):
        return "Transacted"

    elif group == 38 and (
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or
        row['EncounterInsuranceBalance U'] == 0
    ):
        return "Transacted"

    elif group == 39 and (
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
        row['Current Financial Class'] != row['CurrentFinancialClass U'] 
    ):
        return "Transacted"

    elif group == 40 and (
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or
        row['EncounterOpenBalance U'] != row['EncounterOpenBalance'] or
        row['CurrentPPC'] != row['Current Payer Plan Code']
    ):
        return "Transacted"

    elif group == 41 and (
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or
        row['CurrentMajoryPayer'] != row['Current Major Payor']
    ):
        return "Transacted"
        
    elif group == 42 and (
        row['EncounterInsuranceBalance U'] != row['Encounter Insurance Balance'] or
        row['CurrentPPC'] != row['Current Payer Plan Code']
    ):
        return "Transacted"
    
        
        
    return "Not Transacted"

# Apply to DataFrame
Final_df['Validation'] = Final_df.apply(validate_row, axis=1)
```
